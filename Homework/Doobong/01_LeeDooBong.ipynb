```
import matplotlib.pyplot as plt
import numpy as np
from math import *
```

```
m = 1 # mass of particle
k = 0.1 # 용수철상수
t = 0.0 # time 
v = 0.0 # velocity
x = x_0 + 0.1 # position
x_0 = 1.0 # initial position
# a = -9.8 # accelation
```

```
dt =0.1
x_list=[x,]
t_list=[t,]
v_list=[v,]
for i in range(10000):
  t= t + dt
  x = x + v*dt
  a = -k*(x-x_0)/m - 0.01*v # F = ma
  # a = -k*(x-x_0)/m # F = ma
  v = v + a*dt
    

  # if x < 0 : break

  x_list.append(x)
  t_list.append(t)
  v_list.append(v)
```
