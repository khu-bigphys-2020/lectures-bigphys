```
m = 1 # this is mass of particle
t = 0.0 # time
x = 1.0 # position
x0 = 1.0 # default length of the spring
x = x0 + 0.1
v = 0.0 # velocity
k = 0.1 # spring constant
# a = -9.8 # accel
```

```
dt = 0.1
tlist = [t,]
xlist = [x,]
vlist = [v,]

for i in range(15000):
    
  t = t + dt
  x = x + v*dt
  # a = -k*(x-x0)/m # F=ma
  a = -k*(x-x0)/m - 0.01*v
  v = v + a*dt

  # if x < 0: break

  tlist.append(t)
  xlist.append(x)
  vlist.append(v)
  
```
