```
m = 1 # this is mass of particle
t = 0.0 # time
x1 = 1.0 # x1 position
x2 = 2.0 # x2 position
x3 = 3.0 # x3 position
x0 = 1.0 # default length of the spring
x1 = x0 + 0.1
x2 = 2*x0 + 0.1
x3 = 3*x0 + 0.1
v1 = 0.0 # velocity
v2 = 0.0
v3 = 0.0
a1 = 0.0
a2 = 0.0
a3 = 0.0
k = 0.1 # spring constant
# a = -9.8 # accel
```

```
dt = 0.1
tlist = [t,]
x1list = [x1,]
x2list = [x2,]
x3list = [x3,]
v1list = [v1,]
v2list = [v2,]
v3list = [v3,]

for i in range(4000):
      
  t = t + dt
  x1 = x1 + v1*dt
  x2 = x2 + v2*dt
  x3 = x3 + v3*dt
  # a = -k*(x1-x0)/m # F=ma
  a1 = (-k*(x1-x0) + k*(x2-x1-x0))/m - 0.01*v1
  a2 = (-k*(x2-x1-x0) + k*(x3-x2-x0))/m - 0.01*v2
  a3 = (-k*(x3-x2-x0) -k*(x2-2*x0))/m - 0.01*v3
  v1 = v1 + a1*dt
  v2 = v2 + a2*dt
  v3 = v3 + a3*dt


  # if x < 0: break

  tlist.append(t)
  x1list.append(x1)
  x2list.append(x2)
  x3list.append(x3)
  v1list.append(v1)
  v2list.append(v2)
  v3list.append(v3)
  
```
