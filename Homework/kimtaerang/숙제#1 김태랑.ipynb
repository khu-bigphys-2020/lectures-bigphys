```
import matplotlib.pyplot as plt
import random
# 1000회 반복
#양쪽 옆에 벽이있는 상태에서 입자 10개를 용수철로 연결
dt = 0.1
k =0.1
m =1.0
t= 0.0
xlist =[] 
for i in range(10):
    xi_list = []
    for j in range(1):
        xi_list.append(10*(i+1)) # 각 용수철 길이 10
    xlist.append(xi_list)
print(xlist) # 입자위치를 담을 리스트

vlist =[[5],] # 시작 속도 첫 번째 입자만 5
for i in range(10):
    vi_list = []
    for j in range(1):
        vi_list.append(0)
    vlist.append(vi_list)
print(vlist) # 속도를 담을 리스트

tlist = [t,]
print(tlist)
for j in range(1,1000):
    for i in range(len(xlist)):
        if i == 0: # 첫 번째 입자의 운동 -> 오른쪽과 자신만 고려
            xlist[0].append(xlist[0][j-1]+vlist[0][j-1]*dt)
            a = (-k*(xlist[0][j-1]-xlist[0][0])+k*(xlist[1][j-1]-xlist[1][0]-(xlist[0][j-1]-xlist[0][0])))/m
            vlist[0].append(vlist[0][j-1]+a*dt)
        if 0< i < 9 : # 중간 입자들의 운동 -> 양쪽의 입자의 영향을 받음
            a = (-k*(xlist[i][j-1]-xlist[i-1][j-1])+k*(xlist[i+1][j-1]-xlist[i][j-1]))/m
            vlist[i].append(vlist[i][j-1]+a*dt)
            xlist[i].append(xlist[i][j-1]+vlist[i][j-1]*dt)
        if i == 9 : #마지막 입자의 운동
            a = (-k*(xlist[i][j-1])-k*(xlist[i][j-1]-xlist[i-1][j-1]))/m
            vlist[i].append(vlist[i][j-1]+a*dt)
            xlist[i].append(xlist[i][j-1]+vlist[i][j-1]*dt)
    t= t+dt
    tlist.append(t)
plt.figure()
for i in range(10):
  plt.plot(tlist,xlist[i])
plt.figure()
for i in range(10):
  plt.plot(tlist,vlist[i])

plt.figure()
for i in range(10):
  plt.plot(xlist[i],vlist[i])
```
