```
m = 1 # this is mass of the particle
t = 0.0
x0 = 1.0 #default length of the spring
x = x0+ 0.1 # position
v = 1.0
k =0.1 # spring const
#a = -9.8 # accel
```

```
dt = 0.1
tlist = [t,]
xlist = [x,]
vlist = [v,]
for i in range(10000):
  t = t+ dt
  x = x + v*dt
  a = -k*(x-x0)/m - 0.01 *v
  #a = -k*(x-x0)/m #F =ma
  v = v +a*dt
  tlist.append(t)
  xlist.append(x)
  vlist.append(v)
```

```
plt.plot(tlist,xlist)
```

```
plt.plot(tlist,vlist)
```
