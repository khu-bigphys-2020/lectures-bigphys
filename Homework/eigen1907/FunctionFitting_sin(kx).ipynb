```
import random
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```
x = []
y = []
for i in range(100):
    x0 = random.uniform(-2*pi, 2*pi)
    y0 = sin(0.5 * x0) + random.gauss(-0.1, 0.1)
    x.append(x0)
    y.append(y0)

print(len(x))
print(len(y))
```

```
plt.plot(x, y, '.')
plt.show()
```

```
def f(x, k = 1):
    return sin(k * x)
```

```
bestLoss = 1e9
bestK = 0
dt = 0.001
lossHistory = []
kHistory = []
k = 0.9
for i in range(5000):
    kGrad = [2 * (y[i] - f(x[i], k)) * x[i] * cos(k*x[i]) for i in range(len(y))]
    k += sum(kGrad) / len(kGrad) * dt

    reses = [(y[i] - f(x[i], k))**2 for i in range(len(y))]
    loss = sqrt(sum(reses) / len(reses))
    lossHistory.append(loss)
    kHistory.append(k)
    
    if loss < bestLoss:
        bestK = k
        bestLoss = loss

print(bestLoss, bestK)
plt.plot(lossHistory, '-o')
plt.show()


plt.plot(kHistory, '-o')
plt.show()
```

```
plt.plot(x, y, '.')
xForPlot = np.linspace(-2*pi, 2*pi, 1000)
yForPlot = [f(xForPlot[i], bestK) for i in range(len(xForPlot))]
plt.plot(xForPlot, yForPlot, '-')
plt.show()
```
