```
import numpy as np
import matplotlib.pyplot as plt

#initial condition
m = 1.0
t = 0.0
x_0 = 1.0
x = x_0 + 0.1
v = 0.0
k = 0.1

dt = 0.1

tlist = [t,]
xlist = [x,]
vlist = [v,]

for i in range(10000):
  t = t + dt
  x = x + v*dt
  a = -k*(x - x_0)/m - 0.01*v
  v = v + a*dt
  
  tlist.append(t)
  xlist.append(x)
  vlist.append(v)

plt.plot(tlist, xlist)
plt.show()

plt.plot(xlist, vlist)
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

print("몇개의 오실레이터 입자를 만들까요?")
oscilNum = int(input())

deltaT = 0.001
repetation = 100000

tTable = []
xTable = [] #first row: initial location, each column: each particle 
vTable = []
particleDistance = 10
k = 0.1
m = 1

#initial condition
tTable.append(0)
for i in range(oscilNum):
  xLine = [particleDistance*i, ]
  vLine = [0, ]
  xTable.append(xLine)
  vTable.append(vLine)
vTable[oscilNum - 1] = [10, ] # first particle's v is not zero

print(xTable)
print(vTable)
print(tTable)

for i in range(1, repetation):
  #first particle
  vTable[0].append(vTable[0][i-1] + k/m*(xTable[1][i-1] - 2*xTable[0][i-1] - particleDistance)*deltaT)
  xTable[0].append(xTable[0][i-1] + vTable[0][i-1]*deltaT)
  #middle particle
  for j in range(1, oscilNum-1):
    vTable[j].append(vTable[j][i-1] + k/m*(xTable[j+1][i-1] - 2*xTable[j][i-1] + xTable[j-1][i-1])*deltaT)
    xTable[j].append(xTable[j][i-1] + vTable[j][i-1]*deltaT)
  #last particle
  vTable[oscilNum-1].append(vTable[oscilNum-1][i-1] - k/m*(xTable[oscilNum-1][i-1] - xTable[oscilNum-2][i-1] - 10)*deltaT)
  xTable[oscilNum-1].append(xTable[oscilNum-1][i-1] + vTable[oscilNum-1][i-1]*deltaT)
  
  tTable.append(tTable[i-1] + deltaT)

for i in range(oscilNum):
  plt.plot(tTable, xTable[i])
plt.show()

for i in range(oscilNum):
  plt.plot(tTable, vTable[i])
plt.show()

for i in range(oscilNum):
  plt.plot(xTable[i], vTable[i])
plt.show()
```
