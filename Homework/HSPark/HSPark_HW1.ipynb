```
from math import *
import matplotlib.pyplot as plt
```

```
m = 1
t = 0.0
k = 0.1
x0 = 0
x1 = 10
x2 = 20
x3 = 30
x4 = 40
x5 = 50
x6 = 60
x7 = 70
x8 = 80
x9 = 90
x10 = 100 +7
x11 = 110

v1 = 0
v2 = 0
v3 = 0
v4 = 0
v5 = 0
v6 = 0
v7 = 0
v8 = 0
v9 = 0
v10 = 0
dt = 0.1
```

```
tlist = [t,]
x1list = [x1,]
v1list = [v1,]
x2list = [x2,]
v2list = [v2,]
x3list = [x3,]
v3list = [v3,]
x4list = [x4,]
v4list = [v4,]
x5list = [x5,]
v5list = [v5,]
x6list = [x6,]
v6list = [v6,]
x7list = [x7,]
v7list = [v7,]
x8list = [x8,]
v8list = [v8,]
x9list = [x9,]
v9list = [v9,]
x10list = [x10,]
v10list = [v10,]

for i in range(10000):    
    t = t + dt

    x1 = x1 + v1*dt
    x2 = x2 + v2*dt
    x3 = x3 + v3*dt
    x4 = x4 + v4*dt
    x5 = x5 + v5*dt
    x6 = x6 + v6*dt
    x7 = x7 + v7*dt
    x8 = x8 + v8*dt
    x9 = x9 + v9*dt
    x10 = x10 + v10*dt


    a1 = -k*(x1 - x0)/m + k*(x2 - x1)/m
    #a1 =  k*(x2 - x1)/m
    a2 = -k*(x2 - x1)/m + k*(x3 - x2)/m
    a3 = -k*(x3 - x2)/m + k*(x4 - x3)/m
    a4 = -k*(x4 - x3)/m + k*(x5 - x4)/m
    a5 = -k*(x5 - x4)/m + k*(x6 - x5)/m
    a6 = -k*(x6 - x5)/m + k*(x7 - x6)/m
    a7 = -k*(x7 - x6)/m + k*(x8 - x7)/m
    a8 = -k*(x8 - x7)/m + k*(x9 - x8)/m
    a9 = -k*(x9 - x8)/m + k*(x10 - x9)/m
    a10 = -k*(x10 - x9)/m + k*(x11 - x10)/m
    #a10 = -k*(x10 - x9)/m 


    v1 = v1 + a1*dt
    v2 = v2 + a2*dt
    v3 = v3 + a3*dt
    v4 = v4 + a4*dt
    v5 = v5 + a5*dt
    v6 = v6 + a6*dt
    v7 = v7 + a7*dt
    v8 = v8 + a8*dt
    v9 = v9 + a9*dt
    v10 = v10 + a10*dt


    tlist.append(t)
    x1list.append(x1)
    v1list.append(v1)
    x2list.append(x2)
    v2list.append(v2)
    x3list.append(x3)
    v3list.append(v3)
    x4list.append(x4)
    v4list.append(v4)
    x5list.append(x5)
    v5list.append(v5)
    x6list.append(x6)
    v6list.append(v6)
    x7list.append(x7)
    v7list.append(v7)
    x8list.append(x8)
    v8list.append(v8)
    x9list.append(x9)
    v9list.append(v9)
    x10list.append(x10)
    v10list.append(v10)
```
