```
import numpy as np
import math
import time
import matplotlib.pyplot as plt

#Initial Constant
m = 0.5 #질량은 0.5kg
k = 10 #용수철 상수는 10N/m
l = 1 #용수철의 길이는 1m

#용수철을 물리적으로 풀어, 이미 초기조건이 주어진 용수철의 시간에 따른 변위를 알고 있다고 가정한다. 
w = math.sqrt(k/m) #각속도는 k/m의 루트
def acc(x):
  return -x*(w**2) #가속도는 마이너스 변위 곱하기 각속도의 제곱
def pos(t, d=0):
  return d*math.cos(w*t) #변위는 최대변위 곱하기 wt의 cos

#시간은 0초에서 시작하여 60초동안 움직이고, 단위시간을 0.01초로 가정하자
time_range = np.arange(0, 60, 0.1)

# x=0 위치에 용수철 하나가 고정되어 있고, d_1 만큼 용수철을 늘려, x_1의 위치를 ㅣ+d_1에 위치 시켰다고 생각하자. 
def x_1(t, d_1=0):
  return l+pos(t,d_1)

#0.5m를 잡아당긴다고 가정하면,
stop = np.array([])
for t in range(len(time_range)):
  stop = np.append(stop, x_1(time_range[t],0.5))
"""
#계산한 결과를 고정된 축에대해 보면
for data in range(len(stop)):
  plt.axis([0,2,-1,1])
  draw = plt.scatter(stop[data], 0)
  plt.pause(1)
  draw.remove()
"""
#이제 두 번째 용수철이 첫 번째 용수철에 붙은 입자에 붙어 있다고 생각하자. 문제는 첫번째 입자가 가속하기 때문에 두번째 용수철에 힘이 작용된다 
t=0
dt = 0.1
v1 = 0
v2 = 0
tlist = [t,]
x1list = [x1,]
v1list = [v1,]
x2list = [x2,]
v2list = [v2,]
#t=0에서 시작하고, 시간간격은 0.1, 두 입자모두 정지  상태에서 운동하고, 첫번째 입자만 0.5 잡아당긴다고 가정한다
x1 = l + 0.5
x2 = x1 + l
for t in range(len(time_range)):
    t = t + dt

    x1 = x1 + v1*dt
    a1 = -k*(x1-l)/m + k*(x2-x1-l)/m
    v1 = v1 + a1*dt

    x2 = x2 + v2*dt
    a2 = -k*(x2-x1-l)/m -k*(x1-l)/m
    v2 = v2 + a2*dt

    tlist.append(t)
    x1list.append(x1)
    v1list.append(v1)
    x2list.append(x2)
    v2list.append(v2)

print(x1list)

#발산해버린다.



```
