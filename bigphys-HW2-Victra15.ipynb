# 빅데이터 물리학 과제2

```python
import math
import matplotlib.pyplot as plt
import time, sys
from tqdm import notebook
import csv
import pandas
import numpy
```

```python
size = int(input(prompt="input size of space : "))
V = float(input(prompt="input the electric potential of above side : "))
```

```python
electric_potential = numpy.zeros((size + 2, size + 2), float)
electric_potential[size + 1][0:] = V
for n in notebook.tqdm(range(3000)):
    for y in range(size, 0, -1):
        for x in range(size, 0, -1):
            electric_potential[y][x] = (electric_potential[y + 1][x] + electric_potential[y - 1][x] + electric_potential[y][x + 1] + electric_potential[y][x - 1]) / 4
```

```python
print(electric_potential)
```

```python
f = open("electric_potential.csv", 'w')
wr = csv.writer(f)
for y_potential in electric_potential:
        wr.writerow(y_potential)
f.close()
```
