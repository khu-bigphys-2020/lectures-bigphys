```
m = 1 # this is mass of the particle
t = 0.0 # time
x0 = 1.0 # default length of the spring
k = 0.1 # spring constant
x = x0 + 0.1 # position
v = 0.0 # velocity
#a = -9.8 # accel
```

```
dt = 0.1
tlist = [t,]
xlist = [x,]
vlist = [v,]
for i in range(10000):
  t = t + dt
  x = x + v*dt
  a = -k*(x-x0)/m - 0.01*v
  #a = -k*(x-x0)/m #F = ma 
  #a - -9.8
  v = v + a*dt

  #if x <= 0: break

  tlist.append(t)
  xlist.append(x)
  vlist.append(v)
```

```
print(tlist, xlist, vlist)
```

```
import matplotlib.pyplot as plt
```
